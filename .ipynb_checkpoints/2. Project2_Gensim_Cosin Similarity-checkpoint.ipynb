##  Trường hợp 1: Khi người dùng chọn 1 sản phẩm bất kỳ

```python
def get_recommendations (n,products_gem,cosine_sim=similarities) :
  products_gem_re = [[re.sub('[0-9]+','', e) for e in text] for text in products_gem] # số
  products_gem_re = [[t.lower() for t in text if not t in ['', ' ', ',', '.', '...', '-',':', ';', '?', '%', '_%' , '(', ')', '+', '/', 'g', 'ml']] for text in  products_gem_re] # ký tự đặc biệt
  products_gem_re = [[t for t in text if not t in stop_words] for text in products_gem_re] # stopword
  dictionary = corpora.Dictionary(products_gem_re)
  dictionary.token2id 
  corpus = [dictionary.doc2bow(text) for text in products_gem_re]
  feature_cnt = len(dictionary.token2id)
  # Use TF-IDF Model to process corpus, obtaining index
  tfidf = models.TfidfModel(corpus)  # ma trah day dac
# tính toán sự tương tự trong ma trận thưa thớt
  index = similarities.SparseMatrixSimilarity(tfidf[corpus],  # dua cac copus vao tfidf
                                            num_features = feature_cnt)
  data_result = pd.DataFrame(index[tfidf[corpus]]) 
  data_result[n].sort_values(ascending=False)[1:6].index[:5]
  result=df.iloc[data_result[n].sort_values(ascending=False)[1:6].index[:5]]
  import pickle
  with open('corpus_gensim_th1.plk', 'wb') as f:
    pickle.dump(corpus , f)
  import pickle
  with open('tfidf_gensim_th1.plk', 'wb') as f:
    pickle.dump(tfidf , f)
  return result
```

```python
n=int(input('User chọn mã sản phẩm thứ:'))
print(df.iloc[n])
```

```python
get_recommendations(n,products_gem)
```

##  Trường hợp 2: Khi người dùng nhập thông tin liên quan


## User nhập Thông tin sản phẩm

```python
def recommender(product):
  # product='quần đùi'
  product_1 = product.lower().split()
  kw_vector_1 = dictionary.doc2bow(product_1)
  sim = index[tfidf[kw_vector_1]]
  series = pd.Series(sim) 
  series=pd.DataFrame(series.sort_values(ascending=False))[:5].index
  result=df.iloc[series]
  return result
```

```python
sp = input('User nhập info sản phẩm :' )
recommender(sp)
```

```python
def get_recommendations (product_id, cosine_sim=cosine_similarities):
  print(pd.DataFrame(df[df['product_id']==product_id]))
  indices=pd.Series(df.index,index=df['product_id']).drop_duplicates()
  idx=indices[product_id]
  sim_score=enumerate(cosine_sim[idx])
  sim_score=sorted(sim_score,key=lambda x:x[1],reverse=True)
  sim_score=sim_score[1:11]
  for i in sim_score:
    print(i)
  sim_index=[i[0] for i in sim_score]
  print(df.product_name.iloc[sim_index])
product_id = int(input("Nhập product ID: "))
get_recommendations (product_id)
```
